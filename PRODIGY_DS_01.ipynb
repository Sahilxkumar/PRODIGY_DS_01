```python
#choosing column to standardise
c_id= data['customer_id']
# StandardScaler expects a 2D array, so we reshape it accordingly
c_id = np.array(c_id).reshape(-1, 1)
#fitting data and transformation
transformed_column = std.fit_transform(c_id)
data['customer_id'] = transformed_column
```

```python
#choosing column to standardise
am= data['amount']
# StandardScaler expects a 2D array, so we reshape it accordingly
am = np.array(am).reshape(-1, 1)
#fitting data and transformation
transformed_column = std.fit_transform(am)
data['amount'] = transformed_column
```

# Logistic regression

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report
```

```python
# Splitting the data into features (a) and target variable (b)
a = data[['merchant_id', 'customer_id', 'amount', 'card_type_n', 'purchase_category_n', 'customer_age']]
b = data['is_fraudulent']
```

```python
# Encoding categorical variables using OneHotEncoder
encoder = OneHotEncoder()
a_encoded = encoder.fit_transform(a[['card_type_n', 'purchase_category_n']])
```

```python
# Concatenating encoded features with numerical features
a_numerical = a[['merchant_id', 'customer_id', 'amount', 'customer_age']]
a_encoded_df = pd.DataFrame(a_encoded.toarray(), columns=encoder.get_feature_names_out(['card_type_n', 'purchase_category_n']))
a_processed = pd.concat([a_numerical, a_encoded_df], axis=1)
```

```python
# Splitting the data into train and test sets
a_train, a_test, b_train, b_test = train_test_split(a_processed, b, test_size=0.2, random_state=42)
```

```python
# Scaling numerical features
scaler = StandardScaler()
a_train_scaled = scaler.fit_transform(a_train)
a_test_scaled = scaler.transform(a_test)
```

```python
# Initializing and training the logistic regression model
log_reg_model = LogisticRegression(max_iter=1000, random_state=42)
log_reg_model.fit(a_train_scaled, b_train)
```

```python
# Predicting on the test set
b_pred = log_reg_model.predict(a_test_scaled)
```

```python
plt.scatter(b_test,b_pred,color='blue')
```

```python
log_reg_model.score(a_test_scaled,b_test)
```

```python
#classification report
print(classification_report(b_test, b_pred))
```
